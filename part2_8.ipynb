```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_kernel(beta,x,t):
    return np.exp(beta*(-np.linalg.norm(x-t)))
```

```python
x = np.asarray([-6,-4,-3,0,2,4])
new_point = np.asarray([1.5])

points = np.random.random(100)
kernel_matrix = [0.0]*points.shape[0]

for i in range(points.shape[0]):
    kernel_matrix[i] = gaussian_kernel(1,points[i],0.5)

plt.scatter(points,kernel_matrix)
```
